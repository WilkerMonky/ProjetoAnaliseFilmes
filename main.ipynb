```python
def obter_relacoes_filme_categoria(df_principal, coluna_id_filme, coluna_generos, df_categoria, coluna_id_categoria, coluna_nome_categoria):
    # Criar uma cópia do DataFrame de filmes e remover linhas com valores nulos
    df_filmes_temp = df_principal.copy().dropna()

    # Lista para armazenar as relações [id_filme, id_categoria]
    lista_relacoes = []

    # Iterar sobre cada linha do DataFrame de filmes
    for i, row in df_filmes_temp.iterrows():
        filme_id = row[coluna_id_filme]
        lista_categorias = string_para_lista(row[coluna_generos])
        
        # Iterar sobre cada categoria e procurar no DataFrame de categorias
        for categoria in lista_categorias:
            categoria_row = df_categoria[df_categoria[coluna_nome_categoria].str.lower() == categoria.lower()]
            if not categoria_row.empty:
                id_categoria = categoria_row.iloc[0][coluna_id_categoria]
                # Adicionar a relação [id_filme, id_categoria] à lista
                lista_relacoes.append([filme_id, id_categoria])

    return lista_relacoes
```
